```python
Age_of_Driver1 = data["Age_of_Driver"].dropna() 
def replace_missing_Age_of_Driver(row):
    
    Age_of_Driver = row["Age_of_Driver"] # zistíme si vek, pohlavie a titul pasažiera
    Sex_of_Driver = row["Sex_of_Driver"]
    Accident_Severity = row["Accident_Severity"]
    if pd.isna(Age_of_Driver):
        return ptable["Age_of_Driver"][(Sex_of_Driver, Accident_Severity)]
    else:
        return Age_of_Driver 

Age_of_Driver2 = data.apply(replace_missing_Age_of_Driver, axis=1) 
p = plt.hist([Age_of_Driver1, Age_of_Driver2])
```

```python
ptable = pd.pivot_table(data, index=["Sex_of_Driver", "Road_Type"], values="Engine_Capacity", aggfunc="median")
ptable
```

```python
Engine_Capacity1 = data["Engine_Capacity"].dropna() 
def replace_missing_Engine_Capacity(row):
    
    Engine_Capacity = row["Engine_Capacity"] 
    Sex_of_Driver = row["Sex_of_Driver"]
    Road_Type = row["Road_Type"]
    if pd.isna(Engine_Capacity):
        return ptable["Engine_Capacity"][(Sex_of_Driver, Road_Type)]
    else:
        return Engine_Capacity 

Engine_Capacity2 = data.apply(replace_missing_Engine_Capacity, axis=1) 
p = plt.hist([Engine_Capacity1, Engine_Capacity2])
```

```python
ptable = pd.pivot_table(data, index=["Road_Type", "Accident_Severity"], values="Age_of_Vehicle", aggfunc="median")
ptable
```

```python
Age_of_Vehicle1 = data["Age_of_Vehicle"].dropna() 
def replace_missing_Age_of_Vehicle(row):
    
    Age_of_Vehicle = row["Age_of_Vehicle"] 
    Road_Type = row["Road_Type"]
    Accident_Severity = row["Accident_Severity"]
    if pd.isna(Age_of_Vehicle):
        return ptable["Age_of_Vehicle"][(Road_Type, Accident_Severity)]
    else:
        return Age_of_Vehicle 

Age_of_Vehicle2 = data.apply(replace_missing_Age_of_Vehicle, axis=1) 
p = plt.hist([Age_of_Vehicle1, Age_of_Vehicle2])
```

9. Rozdeľte dáta na trénovaciu a testovaciu množinu v pomere 70/30. Ako cieľový atribút použite atribút `Accident_Severity`. (2b)

```python
X_mnozina = mnozina.drop('Accident_Severity', axis=1) 
y_mnozina = mnozina['Accident_Severity']

print(X_mnozina.shape) 
print(y_mnozina.shape)


from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X_mnozina, y_mnozina, test_size=0.3, random_state=1)
```

10. Spočítajte pomocou funkcie `SelectKBest` a `mutual_score_info` významosti jednotlivých atribútov pre predikciu v trénovacej množine. Skúste získané informácie využiť pri predspracovaní dát pre niektoré z modelov. (3b)

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif

fs = SelectKBest(score_func=mutual_info_classif, k='all')
fs.fit(X_train, y_train)
```

```python
for i in range(len(fs.scores_)):
	print('Atribut %d: %f' % (i, fs.scores_[i]))
    
    
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.show()
```

11. Natrénujte rôzne klasifikačné modely pre predikciu atribútu `Accident_Severity`. Natrénujte nasledovné modely s prednastavenými parametrami:
* k-najbližších susedov
* Rozhodovacie stromy
* Random forests

Jednotlivé modely otestujte pomocou 10-násobnej krížovej validácie pomocou metriky `accuracy`. 

Pozor - pre jednotlivé modely zvoľte vhodný spôsob predspracovania (možná modifikácia predspracovania v kroku 8). (6b) 

```python
from sklearn.neighbors import KNeighborsClassifier 
model = KNeighborsClassifier()                       
model.fit(X_train, y_train)                           
y_model = model.predict(X_test)   

from sklearn.metrics import accuracy_score,precision_score, recall_score


print(f"Uspesnost (accuracy) modelu: {accuracy_score(y_test, y_model)}") 
print(f"Presnosť (precision) modelu: {precision_score(y_test, y_model)}") 
print(f"Návratnosť (recall) modelu: {recall_score(y_test, y_model)}") 
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import confusion_matrix

dt = DecisionTreeClassifier(max_depth=3)     
dt.fit(X_train, y_train)       
y_dt = dt.predict(X_test)       

from sklearn.metrics import accuracy_score,precision_score, recall_score 

print(f"Presnosť (accuracy) modelu: {accuracy_score(y_test, y_dt)}")
print(f"Presnosť (precision) modelu: {precision_score(y_test, y_dt)}")
print(f"Návratnosť (recall) modelu: {recall_score(y_test, y_dt)}")

cm = confusion_matrix(y_test, y_dt)  
print(cm)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

rf = RandomForestClassifier(n_estimators=10000, max_depth=5)
rf.fit(X_train, y_train)

y_rf = rf.predict(X_test)

print(f"Presnosť (accuracy) modelu: {accuracy_score(y_test, y_rf)}")
cm = confusion_matrix(y_test, y_rf)
print(cm)
```

12. Porovnajte natrénované modely aj pomocou ROC krivky testovacej množine. Identifikujte model, ktorý dáva najlepšie výsledky s prednastavenými parametrami. 
V tomto kroku sa posnažte vyladiť model nájdením najvhodnejších parametrov pomocou `GridSearchCV`. Nájdite a vypíšte najlepšiu kombináciu parametrov. (4b)

```python
from sklearn.metrics import roc_curve,auc # z Scikit learn importujeme funkcie pre výpočet ROC krivky a AUC koeficientu

fpr, tpr, thresholds = roc_curve(y_test, y_model, pos_label=1) 

roc_auc = auc(fpr, tpr) 


plt.title('ROC Krivka') 


plt.plot(fpr, tpr, color='green', label = 'ROC krivka modelu (AUC = %0.2f)' % roc_auc) 

plt.legend(loc = 'lower right') 
plt.plot([0, 1], [0, 1],linestyle='--', color='red') 
plt.xlim([0, 1]) 
plt.ylim([0, 1]) 
plt.ylabel('True Positive Rate') 
plt.xlabel('False Positive Rate') 
plt.show() 
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV 


k_range = list(range(1, 50))
print(k_range)
```

```python
param_grid = dict(n_neighbors=k_range)
print(param_grid)
```

```python
grid = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, scoring='accuracy') 
grid.fit(X_train, y_train)
```

```python
print("Najlepšie parametre sú:")
print()
print(grid.best_params_)
print()
print(grid.best_score_)
```
